```python
# XY to latlong then writerows to csv0
# writing json file into csv and drawing usable columns
import json
import csv
import SVY21

#import unicodecsv as csv
with open('hdb-carpark-information.csv') as csv_file:
    data = list(csv.DictReader(csv_file, delimiter=','))
    print(data)
    


with open('hdpCarparksLongLat.csv', 'w+', newline='') as csv_file:
    writer = csv.writer(csv_file)

    writer.writerow(['latitude','longitude'])       
    # module.class
    
    projection = SVY21.SVY21() 
    for row in data:
        print(row)
        north = float(row['y_coord'])
        east = float(row['x_coord'])
        print(east)
        (lat,long) = projection.computeLatLon(north,east)
        row = [lat,long]
        writer.writerow(row)
```
